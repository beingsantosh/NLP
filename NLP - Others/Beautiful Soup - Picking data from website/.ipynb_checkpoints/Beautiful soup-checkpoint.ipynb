### another real time scenario - extracting news from a news website.

```python
result1 = requests.get('https://www.coronavirus.gov/')
content1 = result1.content
```

```python
soup_corona = BeautifulSoup(content1)
```

```python
links = soup_corona.find_all('a')
```

```python
print(f'srno   length of the string---------------- ---   actual text ')
for i,link in enumerate(links):
    if len(link.text.strip())!=0:
        
        print(f'{i:1}  {len(link.text):{10}} {link.text.strip():->{50}}')
```

## all the links

```python
lst = []
print(f'srno   --------text on link------- ---   url ')
for i,link in enumerate(links):
    print(f'{i} {link.text.strip():{25}}  {link.attrs["href"]:{60}}')
    lst.append((i, link.text.strip(),  link.attrs["href"])) 
```

```python
lst # it is a list of tuples which contains text and its corresponding url.
```

# *******************Tables ***************************

### corona data from all the coutries

```python
result_c = requests.get('https://www.worldometers.info/coronavirus/')
```

```python
soup_c= BeautifulSoup(result_c.content)
```

### all links present

```python
links_c = soup_c.find_all('a')
lst = []
print(f'srno   --------text on link------- ---   url ')
for i,link in enumerate(links_c):
    print(f'{i} {link.text.strip():{25}}  {link.attrs["href"]:{60}}')
    lst.append((i, link.text.strip(),  link.attrs["href"])) 
```

### all table present

```python
table_c = soup_c.find_all('table')
for i,tab in enumerate(table_c):
    print(f'{i} ')
    
```

###### only 2 tables present

```python
actual_d=[]
header=[]
for tr in soup_c.find_all('tr'):
    data =[]
    
    for th in tr.find_all('th'):
        data.append(th.text)
        
    if data:
        #print(f'Inserting headers : {",".join(data)}')
        header.append(','.join(data))
        continue
              
            
    for td in tr.find_all('td'):
        data.append(td.text)
        #print(f'Inserting data : {",".join(data)}')
        
    actual_d.append(','.join(data))
```

```python
print(len(actual_d))
print(len(header))
```

```python
actual_d
```

### another way to get the data from table

```python
table = soup_c.find_all('table')
```

```python
# header for table 1
[j.text for j in table[1].find_all('th')]
```

```python
# header for table 2
[j.text for j in table[0].find_all('th')]
```

```python
# content from table 1
table_rows = table[0].find_all('tr')
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    print(row)
```

```python
# content from table 2
table_rows = table[1].find_all('tr')
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    print(row)
```

```python
data = '''<div class="image">
        <a href="http://www.example.com/eg1">Content1<img  
        src="http://image.example.com/img1.jpg" /></a>
        </div>
        <div class="image">
        <a href="http://www.example.com/eg2">Content2<img  
        src="http://image.example.com/img2.jpg" /> </a>
        </div>'''

soup = BeautifulSoup(data)

for div in soup.findAll('div', attrs={'class':'image'}):
    print(div.find('a')['href'])
    print(div.find('a').contents[0])
    print(div.find('img')['src'])
```
